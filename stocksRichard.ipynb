```python
SPY1 = stockslist['SPY']['Close'][-252:]
SPY3 = stockslist['SPY']['Close'][-252*3:]
SPY5 = stockslist['SPY']['Close'][-252*5:]
SPY10 = stockslist['SPY']['Close'][-252*10:]

SPY1_start = SPY1.iloc[0]
SPY1_end = SPY1.iloc[-1]
SPY3_start = SPY3.iloc[0]
SPY3_end = SPY3.iloc[-1]
SPY5_start = SPY5.iloc[0]
SPY5_end = SPY5.iloc[-1]
SPY10_start = SPY10.iloc[0]
SPY10_end = SPY10.iloc[-1]

SPY1_percent_change = (SPY1_end - SPY1_start) / SPY1_start
SPY3_percent_change = (SPY3_end - SPY3_start) / SPY3_start
SPY5_percent_change = (SPY5_end - SPY5_start) / SPY5_start
SPY10_percent_change = (SPY10_end - SPY10_start) / SPY10_start

SPY1_percent_change = SPY1_percent_change * 100 # convert to a percentage
SPY3_percent_change = (SPY3_percent_change * 100)/3 # convert to a percentage
SPY5_percent_change = (SPY5_percent_change * 100)/5 # convert to a percentage
SPY10_percent_change = (SPY10_percent_change * 100)/10 # convert to a percentage


print("The 1-year percent change for SPY is: {:.2f}%".format(SPY1_percent_change))
print("The 3-year percent change for SPY is: {:.2f}%".format(SPY3_percent_change))
print("The 5-year percent change for SPY is: {:.2f}%".format(SPY5_percent_change))
print("The 10-year return for SPY is: {:.2f}%".format(SPY10_percent_change))
```
